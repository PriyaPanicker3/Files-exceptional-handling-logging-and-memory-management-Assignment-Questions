*Files, exceptional handling, logging and
memory management Questions*

1. What is the difference between interpreted and compiled languages?

Compiled languages translate the entire code into machine code before execution, resulting in faster performance (e.g., C, C++). Interpreted languages like Python execute code line-by-line, making debugging easier but slower in execution.

2. What is exception handling in Python?

Exception handling in Python is a way to handle errors during program execution using try, except, else, and finally blocks, allowing the program to continue or exit gracefully.

3. What is the purpose of the finally block in exception handling?

The finally block is used to execute code regardless of whether an exception occurred or not. It is commonly used for cleanup tasks like closing files or releasing resources.

4. What is logging in Python?

Logging in Python is the process of recording program events or errors using the logging module. It helps in tracking the flow and diagnosing issues in code execution.

5. What is the significance of the __del__ method in Python?

The __del__ method is a destructor in Python, automatically called when an object is about to be destroyed. It's used to perform cleanup operations like closing connections or files.

6. What is the difference between import and from ... import in Python?

import loads the entire module, requiring qualified access (module.function()), while from ... import allows direct access to specific parts of a module.

7. How can you handle multiple exceptions in Python?

You can handle multiple exceptions by writing multiple except blocks or using a single except with a tuple of exceptions, e.g., except (ValueError, TypeError):.

8. What is the purpose of the with statement when handling files in Python?

The with statement automatically manages file opening and closing, ensuring the file is closed properly even if an error occurs.

9. What is the difference between multithreading and multiprocessing?

Multithreading runs multiple threads in the same process (sharing memory), while multiprocessing runs separate processes with independent memory, suitable for CPU-bound tasks.

10. What are the advantages of using logging in a program?

Logging helps monitor application behavior, debug issues, track errors, and maintain logs for future reference, making it essential for production systems.

11. What is memory management in Python?

Memory management in Python is the process of allocating and releasing memory during program execution. Python handles this automatically using reference counting and garbage collection.

12. What are the basic steps involved in exception handling in Python?

The basic steps include wrapping code in a try block, catching exceptions using except, optionally using else for code that runs if no exception occurs, and finally for cleanup actions.

13. Why is memory management important in Python?

Efficient memory management prevents memory leaks, improves performance, and ensures that programs run smoothly, especially for large-scale or long-running applications.

14. What is the role of try and except in exception handling?

The try block contains code that may raise an exception, and the except block handles specific or general exceptions, preventing the program from crashing.

15. How does Python's garbage collection system work?

Pythonâ€™s garbage collector automatically removes unused objects to free memory. It uses reference counting and a cyclic garbage collector to detect and clean up memory no longer in use.

16. What is the purpose of the else block in exception handling?

The else block runs code only if no exception occurs in the try block. It helps separate normal execution from error-handling logic.

17. What are the common logging levels in Python?

Common logging levels include DEBUG (detailed info), INFO (general events), WARNING (potential issues), ERROR (errors), and CRITICAL (serious errors requiring immediate attention).

18. What is the difference between os.fork() and multiprocessing in Python?

os.fork() creates a new process by duplicating the current one (Unix only), while the multiprocessing module works across platforms and offers a higher-level API to run separate processes.

19. What is the importance of closing a file in Python?

Closing a file releases system resources, ensures data is written properly to disk, and prevents data corruption or memory leaks.

20. What is the difference between file.read() and file.readline() in Python?

file.read() reads the entire file as a single string, while file.readline() reads one line at a time, which is more memory-efficient for large files.

21. What is the logging module in Python used for?

The logging module records events, errors, and information during program execution. It supports different log levels and allows logs to be saved to files or sent to external systems.

22. What is the os module in Python used for in file handling?

The os module provides functions to interact with the operating system, such as checking if a file exists, renaming, removing files, or navigating directories.

23. What are the challenges associated with memory management in Python?

Challenges include managing large objects, avoiding memory leaks from circular references, and optimizing memory usage in long-running applications.

24. How do you raise an exception manually in Python?

You can raise an exception using the raise keyword followed by an exception type, like raise ValueError("Invalid input").

25. Why is it important to use multithreading in certain applications?

Multithreading is useful in I/O-bound applications to perform multiple tasks concurrently, improving responsiveness and resource utilization.
